## 1. Data Cleaning
- Handle missing values
- Encode categorical variables
- Drop irrelevant columns

```python
# Folder convention: keep raw files under ./data/raw
DATA_DIR = Path("data")
RAW_DIR = DATA_DIR / "raw"
RAW_DIR.mkdir(parents=True, exist_ok=True)

# 1) Load raw CSVs (UCI files use semicolon separator)
mat = pd.read_csv(RAW_DIR / "student-mat.csv", sep=";")
por = pd.read_csv(RAW_DIR / "student-por.csv", sep=";")

# Optional: tag course to know origin
mat["course"] = "math"
por["course"] = "portuguese"

# 2) Save untouched copies (good practice)
mat.to_csv(DATA_DIR / "rawdata_student_mat.csv", index=False)
por.to_csv(DATA_DIR / "rawdata_student_por.csv", index=False)

# 3) Merge (stack) the two datasets
merged = pd.concat([mat, por], axis=0, ignore_index=True)

# 4) Basic type fixes
grade_cols = ["G1", "G2", "G3"]
for c in grade_cols:
    merged[c] = pd.to_numeric(merged[c], errors="coerce")

# 5) Handle missing values (simple policy: drop rows with NA in target or key predictors)
merged = merged.dropna(subset=["G3"])

# 6) Remove exact duplicate rows
merged = merged.drop_duplicates()

# 7) (Optional) Normalize yes/no & binary categories to 0/1 for modeling
yn_cols = ["schoolsup","famsup","paid","activities","nursery","higher","internet","romantic"]
for c in yn_cols:
    if c in merged.columns:
        merged[c] = merged[c].map({"yes":1, "no":0})

# 8) Save cleaned dataset for EDA & modeling
CLEAN_PATH = DATA_DIR / "student_clean.csv"
merged.to_csv(CLEAN_PATH, index=False)

print(f"Cleaned dataset saved to: {CLEAN_PATH.resolve()}")
print(merged.shape)
merged.head()
```

## 2. Exploratory Data Analysis (EDA)
- Summary statistics
- Visualizations (histograms, boxplots, scatterplots, heatmaps)

```python
# =========================
# Exploratory Data Analysis
# =========================

# Imports (safe to re-run)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv("data/student_clean.csv")

# -------------------------------------------------
# 0) Quick sanity check
# -------------------------------------------------
print("Shape of dataset:", df.shape)
print("\nMissing values per column:\n", df.isnull().sum())

# -------------------------------------------------
# 1) Descriptive statistics (numeric + categorical)
# -------------------------------------------------
print("\nSummary statistics:\n")
display(df.describe(include="all"))

# -------------------------------------------------
# 2) Histograms for numeric variables
# -------------------------------------------------
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols].hist(figsize=(15, 10), bins=20)
plt.suptitle("Histograms of Numeric Variables", fontsize=16)
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 3) Boxplots for numeric variables (outlier scan)
# -------------------------------------------------
plt.figure(figsize=(15, 8))
df[numeric_cols].boxplot()
plt.xticks(rotation=90)
plt.title("Boxplots of Numeric Variables")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 4) Correlation heatmap (numeric only)
# -------------------------------------------------
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=["int64", "float64"])
sns.heatmap(numeric_df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 5) Example scatterplot (G1 vs G3)
# -------------------------------------------------
plt.figure(figsize=(6, 4))
sns.scatterplot(x="G1", y="G3", data=df, alpha=0.5)
plt.title("First Period Grade (G1) vs Final Grade (G3)")
plt.xlabel("G1 (0–20)")
plt.ylabel("G3 (0–20)")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 6) Boxplots of G3 by key categorical variables
# -------------------------------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="sex", y="G3", data=df)
plt.title("Final Grade (G3) by Sex")
plt.xlabel("Sex (F/M)")
plt.ylabel("Final Grade (0–20)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="internet", y="G3", data=df)
plt.title("Final Grade (G3) by Internet Access")
plt.xlabel("Internet Access (yes/no)")
plt.ylabel("Final Grade (0–20)")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 7) Basic summaries for categorical variables
# -------------------------------------------------
for col in df.select_dtypes(include=["object"]).columns:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts(dropna=False, normalize=True).head())

# ============================================================
# RESEARCH-QUESTION REFINEMENT: FACTOR ANALYSIS FOCUS
# Targeted analysis on Absences, Study Time, Family Background
# ============================================================

# -------------------------------------------------
# 8) Absences: distribution
# -------------------------------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["absences"], bins=20, kde=False)
plt.title("Distribution of Student Absences")
plt.xlabel("Number of Absences")
plt.ylabel("Count of Students")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 9) Absences: relationship with G3
#    a) Scatterplot Absences vs G3
#    b) G3 by Absence quartiles
# -------------------------------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="absences", y="G3", data=df, alpha=0.5)
plt.title("Absences vs Final Grade (G3)")
plt.xlabel("Number of Absences")
plt.ylabel("Final Grade (0–20)")
plt.tight_layout()
plt.show()

# Quartile bins for absences (labels ensure ordered categories)
abs_q = pd.qcut(df["absences"], q=4, duplicates='drop')
plt.figure(figsize=(6,4))
sns.boxplot(x=abs_q, y="G3", data=df)
plt.title("Final Grade (G3) by Absence Quartiles")
plt.xlabel("Absence Quartiles")
plt.ylabel("Final Grade (0–20)")
plt.tight_layout()
plt.show()

# Numeric correlation
corr_abs_g3 = df[["absences", "G3"]].corr().iloc[0,1]
print(f"Correlation between Absences and Final Grade (G3): {corr_abs_g3:.3f}")

# -------------------------------------------------
# 10) Study time: relationship with G3
#     (studytime is coded 1–4; higher means more weekly study)
# -------------------------------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="studytime", y="G3", data=df, order=sorted(df["studytime"].dropna().unique()))
plt.title("Final Grade (G3) by Study Time Level")
plt.xlabel("Study Time (1:<2h, 2:2–5h, 3:5–10h, 4:>10h)")
plt.ylabel("Final Grade (0–20)")
plt.tight_layout()
plt.show()

# Optional: jittered stripplot overlay for distribution insight
plt.figure(figsize=(6,4))
sns.stripplot(x="studytime", y="G3", data=df, alpha=0.35)
plt.title("Distribution of G3 by Study Time Level")
plt.xlabel("Study Time Level")
plt.ylabel("Final Grade (0–20)")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 11) Family background: parents' education vs G3
#     (Medu/Fedu coded 0–4; higher means higher education)
# -------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12,4), sharey=True)
sns.boxplot(x="Medu", y="G3", data=df, ax=axes[0])
axes[0].set_title("Mother's Education (Medu) vs G3")
axes[0].set_xlabel("Medu (0–4)")
axes[0].set_ylabel("Final Grade (0–20)")

sns.boxplot(x="Fedu", y="G3", data=df, ax=axes[1])
axes[1].set_title("Father's Education (Fedu) vs G3")
axes[1].set_xlabel("Fedu (0–4)")
plt.tight_layout()
plt.show()

# Correlations (numeric quick check)
corr_medu_g3 = df[["Medu", "G3"]].corr().iloc[0,1] if "Medu" in df and df["Medu"].dtype != 'O' else np.nan
corr_fedu_g3 = df[["Fedu", "G3"]].corr().iloc[0,1] if "Fedu" in df and df["Fedu"].dtype != 'O' else np.nan
print(f"Correlation Medu–G3: {corr_medu_g3:.3f}")
print(f"Correlation Fedu–G3: {corr_fedu_g3:.3f}")

# -------------------------------------------------
# 12) Optional: quick multi-factor peek (pairwise with G3)
#     Adjust the list based on which factors you finalize.
# -------------------------------------------------
candidate_factors = ["absences", "studytime", "failures"]
available = [c for c in candidate_factors if c in df.columns]
if available:
    sns.pairplot(df[available + ["G3"]], diag_kind="hist")
    plt.suptitle("Pairwise Relationships (Selected Factors + G3)", y=1.02)
    plt.show()
```
